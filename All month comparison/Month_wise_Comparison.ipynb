```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('Jan 2023 CPA Data.xlsx', sheet_name='in', header=2, converters={'Campus': str.strip})
```

```python
df2 = pd.read_excel('February 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df3 = pd.read_excel('March 2023 CPA Data.xlsx',header=2, converters={'Campus': str.strip})
```

```python
df1['Month'] = pd.Timestamp('2023-01-01').strftime('%B')
```

```python
df1.head(5)
```

```python
df2['Month'] = pd.Timestamp('2023-02-01').strftime('%B')
```

```python
df2.head(5)
```

```python
df3['Month'] = pd.Timestamp('2023-03-01').strftime('%B')
```

```python
df3.head(5)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3.columns
```

```python
common_cols = list(set(df1.columns).intersection(df2.columns).intersection(df3.columns))
```

```python
merged_df = pd.merge(df1, df2, on=common_cols, how='outer')
merged_df = pd.merge(merged_df, df3, on=common_cols, how='outer')
```

```python
merged_df
```

```python
merged_df.groupby('Month')['Cancelled?'].value_counts()
```

```python
# Sort the "Month" column in ascending order
merged_df['Month'] = pd.Categorical(merged_df['Month'], categories=['January', 'February', 'March'], ordered=True)

# Group data by month and cancelled status
grouped_df = merged_df.groupby('Month')['Cancelled?'].value_counts()

# Plot bar graph
ax = grouped_df.unstack().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

ax.set_ylabel('Count')
ax.set_xlabel('Month')
plt.show()
```

```python
grouped_df = merged_df.groupby('Month')['Campus'].value_counts()
```

```python
merged_df = merged_df.replace('Campus: Siilicon Valley', 'Campus: Silicon Valley')
```

```python
value_counts = merged_df['Campus'].value_counts()

# Plot bar graph
plt.figure(figsize=(10, 8))
ax = value_counts.plot(kind='bar')
ax.set_ylabel('Count')
ax.set_xlabel('Campus')
ax.set_title('Number of Students by Campus')
plt.xticks(rotation=45, ha='right')

# Add value labels to the bars
for i, count in enumerate(value_counts):
    ax.text(i, count + 1, str(count), ha='center')

plt.tight_layout()
plt.show()
```

```python
# Group data by month and campus
grouped_df = merged_df.groupby(['Month', 'Campus']).size().unstack()

# Plot bar graph
plt.figure(figsize=(12, 6))
ax = grouped_df.plot(kind='bar', stacked=True)
ax.set_ylabel('Count')
ax.set_xlabel('Month')
ax.set_title('Distribution of Campuses Month-wise')

plt.legend(title='Campus', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
