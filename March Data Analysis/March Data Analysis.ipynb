```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('March 2023 CPA Data.xlsx',header=2, converters={'Campus': str.strip})
```

```python
cols = ['Program of Study','Campus','Care Unit','Scheduled Services','Had Appointment?',]
```

```python
for colu in cols:
    column_name = colu
    # Create a new DataFrame that groups the data by the values in the column
    grouped_df = df.groupby(column_name).size().reset_index(name='counts')
    
    # Set the plot style
    plt.style.use('ggplot')
    
    plt.figure(figsize=(20,6))

    # Create a bar plot of the counts
    bars = plt.bar(grouped_df[column_name], grouped_df['counts'])
    
    # Set the x-axis label
    plt.xlabel(column_name)

    # Set the y-axis label
    plt.ylabel('Count')

    plt.title('Counts')

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + 0.35, yval + 0.5, yval)
    
    # Show the plot
    plt.show()
```
