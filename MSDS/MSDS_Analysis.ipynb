```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("MSDS_Sumukh.csv")
```

```python
df1
```

```python
df1.columns
```

```python
df1['Invitee Name'].value_counts()
```

```python
name_counts = df1['Invitee Name'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(20, 6))  # Set the figure size

name_counts.plot(kind='bar')  # Plotting the bar graph

plt.title('Frequency of Invitee Names')  # Set the title of the graph
plt.xlabel('Invitee Names')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis

plt.xticks(rotation=45)  # Rotating the x-axis labels for better readability

plt.show()  # Display the plot
```

```python
df1['Response 1'].value_counts()
```

```python
response_counts = df1['Response 1'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(20, 6))  # Set the figure size

response_counts.plot(kind='bar')  # Plotting the bar graph

plt.title('Meeting Purpose Distribution')  # Set the title of the graph
plt.xlabel('Meeting Purpose')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis

plt.xticks(rotation=45)  # Rotating the x-axis labels for better readability

plt.show()  # Display the plot
```

```python
# Convert the "Start Date & Time" column to datetime format
df1['Start Date & Time'] = pd.to_datetime(df1['Start Date & Time'])

# Extract the month and create a new column
df1['Month'] = df1['Start Date & Time'].dt.month

# Print the updated DataFrame
print(df1['Month'])
```

```python
df1['Month'].value_counts()
```

```python
month_counts = df1['Month'].value_counts().sort_index()

# Plotting the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size

month_counts.plot(kind='bar')  # Plotting the bar graph

plt.title('Monthwise Distribution')  # Set the title of the graph
plt.xlabel('Month')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis

plt.xticks(rotation=0)  # Rotating the x-axis labels if needed

plt.show()  # Display the plot
```
