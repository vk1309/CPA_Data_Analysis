```python
df1 = pd.read_excel('April 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df1.columns
```

```python
df1.shape
```

```python
value_counts = df1['Campus'].value_counts()

# Plot bar graph
plt.figure(figsize=(10, 7))
ax = value_counts.plot(kind='bar')
ax.set_ylabel('Count')
ax.set_xlabel('Campus')
ax.set_title('Number of Students by Campus')
plt.xticks(rotation=45, ha='right')

# Add value labels to the bars
for i, count in enumerate(value_counts):
    ax.text(i, count + 0.25, str(count), ha='center')

plt.tight_layout()
plt.show()
```
