```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('April 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df2 = pd.read_excel('May 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df3 = pd.read_excel('CPA Data June 2023.xlsx', converters={'Campus': str.strip})
```

```python
df1['Month'] = pd.Timestamp('2023-06-06').strftime('%B')
```

```python
df1.head(5)
```

```python
df2['Month'] = pd.Timestamp('2023-05-05').strftime('%B')
```

```python
df2.head(5)
```

```python
df3['Month'] = pd.Timestamp('2023-04-04').strftime('%B')
```

```python
df3.head(5)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df3.columns
```

```python
common_cols = list(set(df1.columns).intersection(df2.columns).intersection(df3.columns))
```

```python
common_cols
```

```python
merged_df = pd.merge(df1, df2, on=common_cols, how='outer')
merged_df = pd.merge(merged_df, df3, on=common_cols, how='outer')
```

```python
merged_df.columns
```

```python
merged_df['Location'].value_counts()
```

```python
merged_df['Campus_y'].value_counts()
```

```python
merged_df
```

```python
merged_df = merged_df.replace('MS Computer Science - Align', 'MSCS Computer Science - Align')
```

```python
value_counts = merged_df['Program of Study'].value_counts()

# Plot bar graph
plt.figure(figsize=(10, 7))
ax = value_counts.plot(kind='bar')
ax.set_ylabel('Count')
ax.set_xlabel('Program of Study')
ax.set_title('Program of Study of students')
plt.xticks(rotation=45, ha='right')

# Add value labels to the bars
for i, count in enumerate(value_counts):
    ax.text(i, count + 0.25, str(count), ha='center')

plt.tight_layout()
plt.show()
```
