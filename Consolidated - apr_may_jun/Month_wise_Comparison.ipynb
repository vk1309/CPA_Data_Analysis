```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('April 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df2 = pd.read_excel('May 2023 CPA Data.xlsx', converters={'Campus': str.strip})
```

```python
df3 = pd.read_excel('CPA Data June 2023.xlsx', converters={'Campus': str.strip})
```

```python
df1['Month'] = pd.Timestamp('2023-06-06').strftime('%B')
```

```python
df1.head(5)
```

```python
df2['Month'] = pd.Timestamp('2023-05-05').strftime('%B')
```

```python
df2.head(5)
```

```python
df3['Month'] = pd.Timestamp('2023-04-04').strftime('%B')
```

```python
df3.head(5)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3.columns
```

```python
common_cols = list(set(df1.columns).intersection(df2.columns).intersection(df3.columns))
```

```python
common_cols
```

```python
merged_df = pd.merge(df1, df2, on=common_cols, how='outer')
merged_df = pd.merge(merged_df, df3, on=common_cols, how='outer')
```

```python
merged_df
```

```python
merged_df.groupby('Month')['Cancelled?'].value_counts()
```

```python
# Sort the "Month" column in ascending order
merged_df['Month'] = pd.Categorical(merged_df['Month'], categories=['April', 'May', 'June'], ordered=True)

# Group data by month and cancelled status
grouped_df = merged_df.groupby('Month')['Cancelled?'].value_counts()

# Plot bar graph
ax = grouped_df.unstack().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

ax.set_ylabel('Count')
ax.set_xlabel('Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

merged_df['Month'] = pd.Categorical(merged_df['Month'], categories=['April', 'May', 'June'], ordered=True)

grouped_df = merged_df.groupby('Month')['Admit Term'].value_counts()

plt.rcParams['figure.figsize'] = (8, 8)

ax = grouped_df.unstack().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

ax.set_ylabel('Count')
ax.set_xlabel('Month') 
plt.show()
```
